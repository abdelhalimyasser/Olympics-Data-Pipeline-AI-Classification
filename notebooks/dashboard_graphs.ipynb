```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from dash import Dash, html, dcc
```

```python
discontinued_sports_df = pd.read_csv('discontinued_sports.csv')
discontinued_sports_df.head()
```

```python
medalists_age_df = pd.read_csv('medalists_age.csv')
medalists_age_df.head()
```

```python
olympiad_summry_df = pd.read_csv('olympiad_summry.csv')
olympiad_summry_df.head()
```

```python
sports_summry_df = pd.read_csv('sports_summry.csv')
sports_summry_df.head()
```

```python
fig = px.bar(
    discontinued_sports_df,
    x='contest_year',
    y='number_of_olympics',
    title='Number of Olympics per Contest Year for Discontinued Sports',
    labels={'contest_year': 'Contest Year', 'number_of_olympics': 'Number of Olympics'}
)

fig.update_layout(
    xaxis_tickangle=-45,
    xaxis_title='Contest Year',
    yaxis_title='Number of Olympics'
)

fig.show()
```

```python
discontinued_sports_melted = discontinued_sports_df.melt(
    id_vars=['discipline'],
    value_vars=['gold_medals', 'silver_medals', 'bronze_medals'],
    var_name='medal_type',
    value_name='medal_count'
)

color_map = {
    'gold_medals': 'orange',
    'silver_medals': 'gray',
    'bronze_medals': 'brown'
}
discontinued_sports_melted['color'] = discontinued_sports_melted['medal_type'].map(color_map)

fig = px.bar(
    discontinued_sports_melted,
    x='discipline',
    y='medal_count',
    color='medal_type',
    color_discrete_map=color_map,
    title='Medal Distribution for Discontinued Sports',
    labels={'discipline': 'Discipline', 'medal_count': 'Number of Medals', 'medal_type': 'Medal Type'},
    barmode='group'
)

fig.update_layout(
    xaxis_tickangle=-45,
    legend_title_text='Medal Type',
    xaxis_title='Discipline',
    yaxis_title='Number of Medals'
)

fig.show()
```

```python
fig = px.bar(
    medalists_age_df,
    x='sport',
    y='age',
    title='Age of Medalists by Sport',
    labels={'sport': 'Sport', 'age': 'Age'},
    color='age',
    color_continuous_scale=px.colors.sequential.Plasma
)

fig.update_layout(
    xaxis_tickangle=-45,
    xaxis_title='Sport',
    yaxis_title='Age of Medalist'
)

fig.show()
```

```python
fig = px.histogram(
    medalists_age_df,
    x='age',
    nbins=20,
    title='Age Distribution of Olympic Medalists',
    labels={'age': 'Age of Medalist', 'count': 'Number of Medalists'},
    color_discrete_sequence=px.colors.sequential.Plasma
)

fig.update_layout(
    xaxis_title='Age of Medalist',
    yaxis_title='Number of Medalists'
)

fig.show()
```

```python
season_colors = {
    'Summer': 'red',
    'Winter': 'blue'
}

fig = px.bar(
    olympiad_summry_df.sort_values(by='games_year'),
    x='games_year',
    y='total_events',
    color='season',
    color_discrete_map=season_colors,
    title='Total Events per Olympic Games (by Year and Season)',
    labels={
        'games_year': 'Olympic Year',
        'total_events': 'Number of Total Events',
        'season': 'Olympic Season'
    },
    barmode='group'
)

fig.update_layout(
    xaxis_tickangle=-45,
    xaxis_title='Olympic Year',
    yaxis_title='Number of Total Events',
    legend_title_text='Season'
)

fig.show()
```

```python
events_heatmap_data = olympiad_summry_df.pivot_table(
    index='host_country',
    columns='games_year',
    values='total_events',
    aggfunc='sum'
)

events_heatmap_data = events_heatmap_data.fillna(0)

fig = px.imshow(events_heatmap_data,
                labels=dict(x="Olympic Year", y="Host Country", color="Total Events"),
                x=events_heatmap_data.columns,
                y=events_heatmap_data.index,
                color_continuous_scale=px.colors.sequential.Plasma,
                title="Total Events per Host Country Across Olympic Years")

fig.update_xaxes(side="top")

fig.show()
```

```python
olympiad_summry_melted = olympiad_summry_df.melt(
    id_vars=['host_country', 'host_city', 'games_year'],
    value_vars=['number_of_gold_medals', 'number_of_silver_medals', 'number_of_bronze_medals'],
    var_name='medal_type',
    value_name='medal_count'
)

color_map = {
    'number_of_gold_medals': 'orange',
    'number_of_silver_medals': 'gray',
    'number_of_bronze_medals': 'brown'
}
olympiad_summry_melted['color'] = olympiad_summry_melted['medal_type'].map(color_map)

fig = px.bar(
    olympiad_summry_melted,
    x='host_country',
    y='medal_count',
    color='medal_type',
    color_discrete_map=color_map,
    title='Medal Distribution for Olympiad Sports According to Host Countries',
    labels={'host_country': 'Country', 'medal_count': 'Number of Medals', 'medal_type': 'Medal Type'},
    barmode='group',
    hover_data=['host_city', 'games_year']
)

fig.update_layout(
    xaxis_tickangle=-45,
    legend_title_text='Medal Type',
    xaxis_title='Host Country',
    yaxis_title='Number of Medals'
)

fig.show()
```

```python
comparison_status_df = olympiad_summry_df[[
    'games_year',
    'number_of_gold_medals',
    'total_medals',
    'athlete_with_most_medals',
    'athlete_with_most_medals_country',
    'athlete_with_most_gold_medals',
    'athlete_with_most_gold_medals_country'
]].copy()

comparison_status_df['athlete_match'] = (
    comparison_status_df['athlete_with_most_medals'] ==
    comparison_status_df['athlete_with_most_gold_medals']
)

comparison_status_df['country_match'] = (
    comparison_status_df['athlete_with_most_medals_country'] ==
    comparison_status_df['athlete_with_most_gold_medals_country']
)

def get_comparison_status(row):
    if row['athlete_match'] and row['country_match']:
        return 'Same Athlete & Country'
    elif row['athlete_match']:
        return 'Same Athlete, Different Country'
    elif row['country_match']:
        return 'Different Athlete, Same Country'
    else:
        return 'Different Athlete & Country'

comparison_status_df['comparison_status'] = comparison_status_df.apply(get_comparison_status, axis=1)

status_counts_per_year = comparison_status_df.groupby(['games_year','number_of_gold_medals', 'total_medals', 'comparison_status', 'athlete_with_most_medals', 'athlete_with_most_medals_country']).size().reset_index(name='count')

comparison_colors = {
    'Same Athlete & Country': 'mediumseagreen',
    'Different Athlete, Same Country': 'mediumblue',
    'Different Athlete & Country': 'red'
}

fig = px.bar(
    status_counts_per_year,
    x='games_year',
    y='count',
    color='comparison_status',
    color_discrete_map=comparison_colors,
    title='Comparison of Top Medalist (Overall) vs. Top Gold Medalist Across Olympic Years',
    labels={
        'games_year': 'Olympic Year',
        'count': 'Number of Olympic Games',
        'comparison_status': 'Comparison Status'
    },
    barmode='stack',
    hover_data={
        'games_year': True,
        'count': True,
        'comparison_status': True,
        'number_of_gold_medals': True,
        'total_medals': True,
        'athlete_with_most_medals': True,
        'athlete_with_most_medals_country': True
    }
)

fig.update_layout(
    xaxis_tickangle=-45,
    xaxis_title='Olympic Year',
    yaxis_title='Number of Olympic Games',
    legend_title_text='Comparison Status'
)

fig.show()
```

```python
medal_trends_df = olympiad_summry_df.melt(
    id_vars=['games_year', 'season'],
    value_vars=['number_of_gold_medals', 'number_of_silver_medals', 'number_of_bronze_medals'],
    var_name='medal_type',
    value_name='medal_count'
)

medal_trends_df = medal_trends_df.sort_values(by='games_year')

medal_type_colors = {
    'number_of_gold_medals': 'gold',
    'number_of_silver_medals': 'silver',
    'number_of_bronze_medals': 'brown'
}

fig = px.line(
    medal_trends_df,
    x='games_year',
    y='medal_count',
    color='medal_type',
    color_discrete_map=medal_type_colors,
    title='Trends of Olympic Medals (Gold, Silver, Bronze) Over Time',
    labels={
        'games_year': 'Olympic Year',
        'medal_count': 'Number of Medals',
        'medal_type': 'Medal Type'
    },
    hover_data=['season']
)

fig.update_layout(
    xaxis_title='Olympic Year',
    yaxis_title='Number of Medals',
    legend_title_text='Medal Type'
)

fig.show()
```

```python
fig = px.scatter(
    olympiad_summry_df,
    x='total_events',
    y='total_medals',
    color='season',
    hover_name='host_city',
    hover_data=['games_year', 'host_country', 'total_events', 'total_medals', 'season'],
    title='Total Events vs. Total Medals by Olympic Year and Season',
    labels={
        'total_events': 'Number of Total Events',
        'total_medals': 'Total Number of Medals',
        'season': 'Olympic Season'
    },
    size='total_medals',
    color_discrete_map={'Summer': 'red', 'Winter': 'blue'}
)

fig.update_layout(
    xaxis_title='Number of Total Events',
    yaxis_title='Total Number of Medals',
    legend_title_text='Olympic Season'
)

fig.show()
```

```python
host_country_counts = olympiad_summry_df['host_country'].value_counts().reset_index()
host_country_counts.columns = ['host_country', 'number_of_times_hosted']

fig = px.bar(
    host_country_counts.sort_values(by='number_of_times_hosted', ascending=False),
    x='host_country',
    y='number_of_times_hosted',
    color='number_of_times_hosted',
    color_continuous_scale=px.colors.sequential.Viridis,
    title='Frequency of Olympic Games Hosted by Country',
    labels={
        'host_country': 'Host Country',
        'number_of_times_hosted': 'Number of Times Hosted'
    }
)

fig.update_layout(
    xaxis_tickangle=-45,
    xaxis_title='Host Country',
    yaxis_title='Number of Times Hosted'
)

fig.show()
```
